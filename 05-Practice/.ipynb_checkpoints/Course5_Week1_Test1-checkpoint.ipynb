```python
df["perday"]=df.milk
```

```python
df["perday"]=df.milk/np.array([monthrange(df.month.apply(pd.to_datetime)[i].year,df.month.apply(pd.to_datetime)[i].day)[1] for i in range(len(df))])
```

```python
df.month.apply(pd.to_datetime)[1].day
```

```python
df
```

```python
df.perday.plot()
```

```python
df["daily_diff12"] = df.perday - df.perday.shift(12)
df["daily_diff1"] = df.daily_diff12 - df.daily_diff12.shift(1)
```

```python
df["daily_diff1"].dropna().plot()
```

```python
raw=df["daily_diff1"].dropna()
```

```python
raw
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(raw.values.squeeze(), lags=50)
```
